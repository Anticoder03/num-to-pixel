#### Convert the above `raw image` to `Std. RGB` image

```python
img = img.convert('RGB')
img
```

#### Convet the above image to `Grayscale`

```python
img_grayscale = img.convert('L') #L stands for Luminance
img_grayscale
```

### Get the size of the image

```python
width, height = img_grayscale.size
print("Width of the image:", width, "px")
print("Height of the image:", height, "px")
```

### Let us transform the image into a square one of 300 X 300 pixels

```python
img_grayscale_resized_300 = img_grayscale.resize((300, 300))
```

```python
img_grayscale_resized_300
```

### Now finally lets get this `28 X 28 pixels` in line with MNIST dataset dimensions

```python
img_grayscale_resized_28 = img_grayscale.resize((28,28)) #resized to 28 by 28 pixels
```

```python
img_grayscale_resized_28
```

#### Show the size of the above image

```python
width1, height1 = img_grayscale_resized_28.size
print("Width of the resized image:", width1, "px")
print("Height of the resized image:", height1, "px")
```

### Now convert the image into pixelated one using array conversion

```python
img_grayscale_resized_28_array = np.array(img_grayscale_resized_28)
img_grayscale_resized_28_array
```

```python
print("Dimension of the pixelated array:", img_grayscale_resized_28_array.ndim)
```
